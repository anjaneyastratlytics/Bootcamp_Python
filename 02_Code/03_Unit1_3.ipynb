```python
def get_file_list(file_path):
    file_list = []
    with open(file_path) as f:
        r = csv.DictReader(f)
        for row in r:
            file_list.append(row)
    print(f"Loaded {len(file_list)} records into list.")
    return file_list
```

```python
def check_duplicates(file_list,field):
    seen = set()
    duplicates = []

    for row in file_list:
        id = row.get(field)
        if id in seen:
            duplicates.append(id)
        else:
            seen.add(id)

    print(f"{len(duplicates)} duplicates found.")
    return duplicates
```

```python
def count_by(file_list, field):
    cat_count = {}
    for row in file_list:
        cat = row.get(field)
        if cat in cat_count:
            cat_count[cat] += 1
        else:
            cat_count[cat] = 0

    print(f"{len(cat_count)} unique categories found.")
    return cat_count
```

```python
def access_jsonl(file_path, field_list):
    with open(file_path) as f:
        record_count = 0
        error_record_lines = []
        for line in f:
            record_count += 1
            try:
                record = json.loads(line)
                output = ""
                for field in field_list:
                    output += f"{field}: {record.get(field)}    "
                print(output)
            except json.JSONDecodeError:
                error_record_lines.append(record_count)
                continue
        print(f"Total records: {record_count}, Errors record lines: {error_record_lines}")
```

```python
def get_sales_qty_by(sales_file_path, field):
    sales_qty_by_field = {}
    with open(sales_file_path) as sales:
        record_count = 0
        error_record_lines = []
        for line in sales:
            record_count += 1
            try:
                record = json.loads(line)
                fld = record.get(field)
                qty = record.get("qty")

                if fld in sales_qty_by_field:
                    sales_qty_by_field[fld] += qty
                else:
                    sales_qty_by_field[fld] = 0
            except json.JSONDecodeError:
                error_record_lines.append(record_count)
                continue
        print(f"Total records: {record_count}, Errors record lines: {error_record_lines}")
    return sales_qty_by_field
```
