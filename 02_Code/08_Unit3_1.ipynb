```python
import psycopg2
import os
from dotenv import load_dotenv
load_dotenv()
```

```python
import logging
logging.basicConfig(
    level=logging.INFO,
    format= "%(asctime)s %(levelname)s %(message)s"
)
```

```python
with psycopg2.connect(
    host = os.getenv("DB_HOST"),
    database = os.getenv("DB_NAME"),
    user = os.getenv("DB_USER"),
    password = os.getenv("DB_PASS"),
    port = 5432
) as conn:
    with conn.cursor() as cursor:
        cursor.execute("SELECT column_name FROM information_schema.columns WHERE table_name = 'dealer' AND table_schema = 'public' ORDER BY ordinal_position;")
        result = cursor.fetchall()
        print(result)
```

```python
def execute_query(query,fetch=None):
    '''Estaablishes Connection with Database, executes Query and returns Result'''
    logging.info("Connecting to Database")
    try:
        with psycopg2.connect(
            host = os.getenv("DB_HOST"),
            database = os.getenv("DB_NAME"),
            user = os.getenv("DB_USER"),
            password = os.getenv("DB_PASS"),
            port = 5432
        ) as conn:
            logging.info("Connection Successful")
            with conn.cursor() as cursor:
                logging.info(f"Executing Query | {query}")
                cursor.execute(query)
                logging.info(f"Query Execution Successful")
                if not fetch:
                    return True
                elif fetch.strip().lower() == "one":
                    result = cursor.fetchone()
                    logging.info(f"Result:\n{result}")
                    return result
                elif fetch.strip().lower() == "all":
                    result = cursor.fetchall()
                    logging.info(f"Result:\n{result}")
                    return result
                else:
                    logging.error(f"Invalid fetch mode: {fetch} | Valid fetch modes: ['one','all',None]")
    except Exception as e:
        logging.error(e)
        return False
```

```python
query = "SELECT * FROM dealer;"
execute_query(query,"one")
```

```python
query = '''CREATE TABLE temp(
temp_id INT,
temp_name VARCHAR(30))'''
execute_query(query)
```
