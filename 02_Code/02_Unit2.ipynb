```python
def normalize_null(value):
    if value is None:
        return None

    value = value.strip()
    
    if value.lower() in {"","na","null","nan"}:
        return None
    
    return value
```

```python
def safe_int(value):
    try:
        return int(value)
    except(ValueError, TypeError):
        return None
    
def safe_float(value):
    try:
        return float(value)
    except(ValueError, TypeError):
        return None
```

```python
with open(root_path + "01_Data/dirty/dealer.csv", newline="") as f_in, open(root_path + "03_Output/clean_dealer.csv", mode="w", newline="") as f_clean, open(root_path + "03_Output/reject_dealer.csv", mode="w", newline="") as f_reject:
    
    reader = csv.DictReader(f_in)
    clean_writer = csv.DictWriter(f_clean, fieldnames=reader.fieldnames)
    clean_writer.writeheader()
    reject_writer = csv.DictWriter(f_reject, fieldnames=reader.fieldnames) 
    reject_writer.writeheader()
    
    clean_count = 0
    reject_count = 0
    
    for row in reader:
        dealer_id =  normalize_null(row.get("dealer_id",None))
        dealer_name =  normalize_null(row.get("dealer_name",None))
        region = normalize_null(row.get("region",None)).upper()
        credit_terms_days =  normalize_null(row.get("credit_terms_days",None))
        
        print(f"Processing dealer {dealer_id}...")
        
        row["dealer_id"] = dealer_id
        row["dealer_name"] = dealer_name
        row["region"] = region
        row["credit_terms_days"] = credit_terms_days
        
        if (dealer_id is None):
            reject_writer.writerows(row)
            print(f"Dealer {dealer_name} has no dealer_id")
            reject_count+=1
        elif (dealer_name is None):
            reject_writer.writerows(row)
            print(f"Dealer {dealer_id} has no dealer_name")
            reject_count+=1
        elif (region not in {"NORTH","SOUTH","EAST","WEST"}):    
            reject_writer.writerows(row)
            print(f"Dealer ({dealer_id}:{dealer_name}) has invalid region {region}")
            reject_count+=1
        elif (credit_terms_days is None) or not (str.isdigit(credit_terms_days)):
            reject_writer.writerows(row)
            print(f"Dealer ({dealer_id}:{dealer_name}) has invalid credit_terms_days {credit_terms_days}")
            reject_count+=1
        else:
            clean_writer.writerow(row)
            print(f"Dealer ({dealer_id}:{dealer_name}) has a valid record.")
            clean_count+=1
    
    print(f"Created Clean file with {clean_count} record(s) and reject file with {reject_count} record(s).")
        
```
